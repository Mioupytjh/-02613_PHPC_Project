```python
from os.path import join
import numpy as np
import matplotlib.pyplot as plt

LOAD_DIR = "/dtu/projects/02613_2025/data/modified_swiss_dwellings/"

with open(join(LOAD_DIR, "building_ids.txt"), "r") as f:
    building_ids = f.read().splitlines()

print(f"Number of buildings: {len(building_ids)}")
print("First 5 IDs:", building_ids[:5])
```

```python

def load_data(load_dir, bid):
    SIZE = 512
    u = np.zeros((SIZE + 2, SIZE + 2))
    u[1:-1, 1:-1] = np.load(join(load_dir, f"{bid}_domain.npy"))
    interior_mask = np.load(join(load_dir, f"{bid}_interior.npy"))
    return u, interior_mask

```

```python
bid = building_ids[0]
u0, interior_mask = load_data(LOAD_DIR, bid)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(u0[1:-1, 1:-1], cmap="inferno")
axes[0].set_title(f"ID {bid} - domain")
axes[0].axis("off")

axes[1].imshow(interior_mask, cmap="gray")
axes[1].set_title(f"ID {bid} - interior mask")
axes[1].axis("off")

plt.tight_layout()
plt.show()
```

```python
sample_ids = building_ids[:4]

fig, axes = plt.subplots(len(sample_ids), 2, figsize=(10, 4 * len(sample_ids)))

for row, bid in enumerate(sample_ids):
    u0, interior_mask = load_data(LOAD_DIR, bid)

    axes[row, 0].imshow(u0[1:-1, 1:-1], cmap="inferno")
    axes[row, 0].set_title(f"ID {bid} - domain")
    axes[row, 0].axis("off")

    axes[row, 1].imshow(interior_mask, cmap="gray")
    axes[row, 1].set_title(f"ID {bid} - interior mask")
    axes[row, 1].axis("off")

plt.tight_layout()
plt.show()
```
